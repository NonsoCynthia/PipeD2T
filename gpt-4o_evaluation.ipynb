```python
def get_completion(prompt, model):  # model="gpt-3.5-turbo", 'gpt-3.5-turbo-16k', 'gpt-3.5-turbo', 'gpt-4'
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0
    )
    return response.choices[0].message.content.strip()
```

```python
import pandas as pd

# Load the CSV file
eval = pd.read_csv('/content/human_eval.csv', encoding='utf-8', delimiter=',')

# Define the instruction
# instruct = '''You are a participant in a human evaluation for texts generated by language models.
# You are provided with a source and a target text with a task to give a yes/no answer if the texts have an omission, addition, incorrect number, and incorrect entities(referring expressions).
# Ignore or do not consider capitalization, lower cased entities and punctuations in both numbers (5.0 or 5, 5,000 or 5000) and texts as an error.
# Please give only a yes/no result and indicate the inconsistencies. Strictly keep to the source information and input structure.'''

instruct = '''You are a participant in a human evaluation for texts generated by language models.
You are provided with a source and a target text with a task to give a yes/no answer if the texts have an omission, addition, incorrect number, and incorrect entities(referring expressions).
Ignore or do not consider capitalization (andrew is same Andrew), lower cased entities and punctuations in both numbers (5.0 is same 5, 5,000 is same 5000) and texts (UK or uk, Dublin, Ireland or Dublin Ireland, meters is same metres) as an error.
Please give only a yes/no result and indicate the inconsistencies. Strictly keep to the answer structure.'''

data = []

# Iterate over each row in the DataFrame starting from the 121st row
for index, row in eval.iloc[:481].iterrows():
    source_text = row['Source Text']
    generated_text = row['Generated Text']
    text = f"Instruction: {instruct}\nSource Text: {source_text}\nGenerated Text: {generated_text}\nOmission:\nAddition:\nIncorrect Number:\nIncorrect Entity:"
    data.append(text)

# Check the length of the data list
print(len(data))

# # Optionally, print the first few entries to verify
# for entry in data[:2]:
#     print(entry)
```

```python
model="gpt-4o"
result_data = []
for prompt in data:
  result = get_completion(prompt, model).replace('\n',' ')
  result_data.append(result)
  # print(result)

with open('result_data333.txt', 'w') as file:
  file.write('\n'.join(result_data))
```
